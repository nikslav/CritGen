```python
import re
from copy import deepcopy
from collections import namedtuple, OrderedDict, defaultdict, Counter
import os
```

```python
os.listdir("../transcripts")
```

```python
with open("../transcripts/EP030_FINAL.txt") as f:
    text = f.read()
```

```python
text_stripped = deepcopy(text).replace("\n", " ")
text_stripped = ' '.join(text_stripped.split())

new_speaker = re.compile("([A-Z]+:)", flags=re.DOTALL)
utterances = [utterance for utterance in new_speaker.split(text_stripped) if utterance!=""]
```

```python
utterance_tuples = []

for i in range(0, len(utterances)-1, 2):
    utterance_tuples.append((i//2, utterances[i], utterances[i+1].strip()))
```

```python
speakers_utterances_dict = defaultdict(list)

for utterances in utterance_tuples:
    speakers_utterances_dict[utterances[1]].append((utterances[0], utterances[2].split()))
```

```python
speakers_utterances_dict["MATT:"]
```

```python
speaker_sequence = [utterance[1][:-1] for utterance in utterance_tuples]
```

```python
speaker_transitions = defaultdict(Counter)
```

```python
for prev, next in zip(speaker_sequence[:-1], speaker_sequence[1:]):
    speaker_transitions[prev][next] += 1
```
